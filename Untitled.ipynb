2) Tokenizing text
Tokenizing the text is bascially breaking the text into smaller units, such as individual words and punctutation and then from there you can make token id's and then token embeddings



Created my own dataset to help go over interview concepts and basic coding principles you should know as a software engineering intern:

```python
with open("understanding_llm_input_data/coding_helper_dataset.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

print("Total number of characters:", len(raw_text))
print(raw_text[:99])
```

Now I want to tokenize and embed this for llm

The following regular expression will spereate on white spaces and punctuation.

```python
import re 

preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)
preprocessed = [item for item in preprocessed if item]
print(preprocessed[:38])

print("Number of tokens:", len(preprocessed))
```
